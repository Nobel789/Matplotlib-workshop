```python
##First generate 100 equaly spaced no from 0 to 10

a1 = np.linspace(0, 10, 100)


# create a plot figure, which act as a container of all plot generated
fig = plt.figure()
#in this plot x1 is x axis  and np.sin(a1)  and np.cos(a1) in yaxies and - and --for line marker

plt.plot(a1, np.sin(a1), '-')
plt.plot(a1, np.cos(a1), '--');
```

 Important Notes:
 Matplot object hierachy, we can think of the Figure object as a box-like container containing one or more Axes. The Axes object contain smaller objects such as tick marks, lines, legends, title and text-boxes.
 Matplot API interface, Matplotlib has two APIs .1) A MATLAB-style state-based interface and "2)a more powerful object-oriented (OO) interface(pyplot interfaces), we work on it

Matplotlib.pyplot is a collection of commands and functions that make Matplotlib behave like MATLAB (for plotting). The MATLAB-style tools are contained in the pyplot (plt) interface.

```python
#State-machine interface
#The state-machine interface of plt  automatically creates figures and axes to achieve the desired plot. 
# so you dont need to create figure and axes in every code run
# Lets try to create plot with title, label and adding ligend
a = np.linspace(0, 10, 50)#create ann array with 50 equal space number in o to 10 range

plt.plot(a, a, label='a1linear')
plt.plot(a, a**2, label='b1quadratic')
plt.plot(a, a**3, label='c1cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("General Plot")

plt.legend()

plt.show()
```

```python
#Formating style of plot
#The default format string is 'b-', which is a solid blue line. 
plt.plot([10,23,45,67], [34,65,76,98])
plt.axis([0, 100, 0, 99])# here we select the range of x-axies and y-axies
plt.show()
```

```python
# to create red line
plt.plot([10,23,45,67], [34,65,76,98],"ro")
plt.axis([0, 100, 0, 150])# here we select the range of x-axies and y-axies
plt.show()
```

```python
#working with different array(think about different attribute in df )
# and compare all array lines in single figure by specifying color
# evenly sampled time at 200ms intervals
t = np.arange(0., 8., 0.2)# its creates an array like 0.  0.2 0.4 0.6

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# object oriented API, why we use, if statemachine interface usually control the hierachy
# we use it dealing with complex plot , its give us more controle figure
# First create figure and ax object
fig = plt.figure()

ax = plt.axes()
```

```python
fig = plt.figure()

ax = plt.axes()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])#for the position and dimensions of the axes. For example:
ax.plot(x2, y2, 'g^')

ax.set_xlabel('x2')
ax.set_ylabel('y2')
ax.set_title('title')

plt.show()
```

```python
# Creating multilineplot in one figure, its very useful for comparing the trend of different variable
x4 = range(1, 9)

plt.plot(x4, [xi*2 for xi in x4])

plt.plot(x4, [xi*5 for xi in x4])

plt.plot(x4, [xi/2.0 for xi in x4])

plt.show()
```

```python
#Barchart,Bar charts display rectangular bars either in vertical or horizontal form. 
#Their length is proportional to the values they represent.
#They are used to compare two or more values., where len (data)=4(0 to 3)
data = [5. , 25. , 50. , 20.]

plt.bar(range(len(data)), data)

plt.show() 
```

```python
# if we want to see the bar graph i horizontaly
plt.barh(range(len(data)), data)
```

```python
#Error Barplot, graphical representation of data that includes both the mean or central value 
#and the variability or uncertainty associated with the data points. 
#It is commonly used to display the variation or error in measurements or experimental data.
x9 = np.arange(0, 4, 0.2)  # Array of x values
y9 = np.exp(-x9)  # Array of y values

e1 = 0.1 * np.abs(np.random.randn(len(y9)))  # Array of error values

plt.errorbar(x9, y9, yerr=e1, fmt='.-')  # Plotting the error bar chart
plt.show()  # Display the chart
```

```python
#stacked Bar chart,visualize the total magnitude and relative proportions of multiple variables within each category.
# using bar plot and spicify the begin or bottom to value by mentioning A. 
A = [15., 30., 45., 22.]

B = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, A, color = 'g')
plt.bar(z2, B, color = 'r', bottom = A)

plt.show()
```

```python
#pie chart, effective way to represent information when we are interested mainly in comparing the wedge against the whole pie
# need two parameter, level or slice name and each slice value
x = [35, 25, 20, 20]

labels = ['shari', 'Bari', 'Gari', 'nari']

plt.pie(x, labels=labels);

plt.show()
```

```python
#Boxplot, distributions of values by showing the median, quartiles, maximum and minimum of a set of values.
y = np.random.randn(50)

plt.boxplot(y)

plt.show();
```

```python
#Area chart, like line graph, instead of plt.plot we use plt.fill_between(x,y)
#Area charts are effective in showing the cumulative sum of values over a given time period or category.
x = range(1, 6)
y = [30, 45, 65, 80, 46]

# Area plot
plt.fill_between(x, y,color='green')
plt.show()
```

```python
## adding legend
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0)

ax.legend(['moderate','maximum','minimum']);
plt.xlabel('This is the X axis')

plt.ylabel('This is the Y axis')
```

```python
#control the color,b blue

#c cyan ,g green,k black,m magenta

#r red,w white, y yellow
x16 = np.arange(1, 5)

plt.plot(x16, 'g')
plt.plot(x16+1, 'k')
plt.plot(x16+2, 'm')

plt.show()
```

```python
# Control line style
#Style abbreviation Style . solid line -- dashed line,-. dash-dot line : dotted line
x16 = np.arange(1, 5)

plt.plot(x16, ':', x16+1, '-.', x16+2, '.')

plt.show()
```

```python
# Handeling axies by specifying value if need with axies() function
#Matplotlib automatically sets the limits of the plot to precisely contain the plotted datasets. 
#Sometimes, we want to set the axes limits ourself. -
x15 = np.arange(1, 5)

plt.plot(x15, x15*1.5, x15, x15*3.0, x15, x15/3.0)

plt.axis()   # shows the current axis limits values

plt.axis([0, 5, -1, 13])#Xmin, Xmax, ymin and y max
```
